```python

# Use this to point to specific past-processed TJ output folders
# (each folder must contain both the '*_cp_masks.tif' masks and their
# matching '*_downs.tiff' intensity images, e.g. OUTPUT_FOLDER\TJ_ch3\<date>)
INPUT_FOLDER = [r'Z:\Mikala\env_test\output\TJ_renaming\TJ_ch3\03222026',
                r'Z:\Mikala\env_test\output\TJ_renaming\TJ_ch3\03232026']
#Use this to point to a folder containing all the subfolders to batch process
# INPUT_FOLDER = 'Z:\Mikala\cell_size_issue\test_folders'

#output folder
OUTPUT_FOLDER = r'Z:\Mikala\env_test\output\TJ_renaming\TJ_ch3'

label = 'TJ'
```

## Process TJ folders

```python
if isinstance(INPUT_FOLDER, list):
     print('Running batch metrics retrieval on specified folders: ', INPUT_FOLDER)
     folderList = [f for f in INPUT_FOLDER if os.path.isdir(f)]
     missingFolders = [f for f in INPUT_FOLDER if not os.path.isdir(f)]
     if missingFolders:
         print('WARNING: the following folders do not exist and will be SKIPPED:')
         for f in missingFolders:
             print('	' + f)

else:
    print('Running batch metrics retrieval on all subfolders in: ', INPUT_FOLDER)
    folderList = [os.path.join(INPUT_FOLDER, f) for f in os.listdir(INPUT_FOLDER) if os.path.isdir(os.path.join(INPUT_FOLDER, f))]

print('Folders queued for processing (' + str(len(folderList)) + '):')
for f in folderList:
    print('	' + f)

if not os.path.exists(OUTPUT_FOLDER):
    os.makedirs(OUTPUT_FOLDER)

for folder in folderList:
    print('Processing folder: ', folder)
    basename = os.path.basename(folder)

    TJstats, TJregions = pt.processTJSegmentationMasks(folder, label)
    pt.exportTJSegmentationCounts(TJstats, OUTPUT_FOLDER, basename)
    pt.exportTJSegmentationPercellMetrics(TJregions, OUTPUT_FOLDER, basename)

print("Done processing")
```
